```python
from sklearn.datasets import fetch_20newsgroups
```

```python
data = fetch_20newsgroups(subset='train')
```

```python
df = pd.DataFrame({'text':data.data, 'target':data.target})
```

```python
df.head()
```

```python
df['target_name']=df['target'].map(lambda i: data.target_names[i])
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
df['char_len'] = df['text'].str.len()
```

```python
df.head(2)
```

```python
df['word_count'] = df['text'].str.split().map(len)
```

```python
all_tokens=[]
```

```python
for row in df['text']:
  tokens = row.lower().split()
  all_tokens+=tokens
```

```python
len(all_tokens)
```

```python
all_tokens[:10]
```

```python
from collections import Counter

freq = Counter(all_tokens)
freq.most_common(10)
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import numpy as np
unique, counts = np.unique(y_train, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
pix = x_train.flatten()
print("min, max :", pix.min(), pix.max())
print("mean, std :", pix.mean(), pix.std())
```

```python
zero_frac = np.mean(pix == 0 )
print(zero_frac*100)
```

```python
prototypes = [x_train[y_train==d].mean(axis=0) for d in range(10)]
```
